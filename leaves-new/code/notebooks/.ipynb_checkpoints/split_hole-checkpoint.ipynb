```python
import numpy as np
from path import Path
import matplotlib.pyplot as plt
from glob import glob
from tqdm import tqdm
import os
import sys
import cv2
sys.path.append('../')
import random
from tool.data_io import get_data_paths,read_image, read_json, save_image, try_create_dir,save_json
from tool.yaml_io import write_to_yaml,read_from_yaml
from tool.plt_utils import plt_samples

def set_labels(label_file):
    labels = read_json(label_file)
    label_ids = set()
    label_names = set()
    for species in labels.keys():
        for (idx, item) in labels[species].items():
            label_ids.add(item['number'])
            label_names.add(item['label'])
    return labels
```

```python
# from scipy.ndimage.measurements import label
import scipy.ndimage.measurements as measure
# This functions gives each object of damage in the mask a unique number.
def label_damages(mask):
    """
    Args:
        mask: the annotated mask in semantic-mask directory, should be the same size as image
    Returns:
        labeled_mask: a labeled mask with n_components
        n_components_: The number of components or objects (damages) in masks.
    """
    structure = np.ones((3, 3), dtype=int)
    labeled_mask, n_components_ = measure.label(mask, structure)
    return labeled_mask, n_components_

def mark_white_hole(label, image, leaf_mask, bound = (0.4,  0.01)):
    labeled_mask, n_components_ = label_damages(label == 1)
    image_copy = image.copy()
    white_holes = np.zeros(label.shape)
    binary_image = np.mean(image, axis = 2)
    img_hsv = cv2.cvtColor(image_copy, cv2.COLOR_BGR2HSV)
    for d_id in range(1, n_components_+1):
        white = img_hsv[labeled_mask == d_id]
        saturation = white[:, 1]
        values = white[:, 2]
        n_s = np.count_nonzero(saturation <= 60)
        n_v = np.count_nonzero(values > 160)
        n_all = len(saturation)
        if n_s / n_all > bound[0] and n_v /n_all > bound[1]:
            white_holes[np.logical_and(labeled_mask == d_id, label == 1)] = 1

    index_black_hole = np.logical_and(label == 1, white_holes == 0)
    image_copy[index_black_hole] = [0,255,0]
    image_copy[white_holes == 1] = [255,0,0]
    return (image_copy,white_holes == 1)

```

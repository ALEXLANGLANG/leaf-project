```python
def get_id2dam(phase, config_file='../../configs/data_to_partition.yml',tile_acts_root='/usr/xtmp/xs75/leaves/experiments/visualization/acts/scores'):
    root, id2paths = get_data_paths(config_file)
    labels = read_json('../../stats/label_file_common_binary.json')
    root_leaves= Path(root)
    root = Path(tile_acts_root)
    id2dam = {}
    for path in tqdm(glob(root/phase/'*/init_s1r95p05_mask
    /*.npy')):
        id_ = os.path.join(*path.split('/')[-2:]).split('.')[0]
        sp, name = id_.split('/')
#         if id_ != 'quercus-bicolor-herbivory/00000745_2082548':
#             continue
#         label = labels[sp]
        label, _ = read_image(root_leaves/id2paths[id_]['label'])
        mask, _ = read_image(root_leaves/id2paths[id_]['leaf_mask'])
        act = np.load(path)
        id2paths[id_]['act'] = path
        prods = sigmoid(act)
        preds = prods > 0.5
        id2dam[id_]={}
        for v in np.unique(label):
            d_name = labels[sp][str(int(v))]['label']
            pred_pos = label[preds]
            pred_neg = label[~preds]
            if d_name not in id2dam[id_].keys():
                id2dam[id_][d_name] = {'pos':(pred_pos==int(v)).sum(), 'neg': (pred_neg==int(v)).sum()}
            else:
                id2dam[id_][d_name]['pos'] += (pred_pos==int(v)).sum()
                id2dam[id_][d_name]['neg'] += (pred_neg==int(v)).sum()
    return id2dam, root, root_leaves, id2paths

def get_id2stats(id2dam):
    id2stats= {}
    for id_ in id2dam:
        fp = tp = fn = 0
        for d in id2dam[id_].keys():
            if d == '__background__':
                fp += id2dam[id_][d]['pos'] if 'pos' in id2dam[id_][d].keys() else 0
            else:
                tp += id2dam[id_][d]['pos'] if 'pos' in id2dam[id_][d].keys() else 0
                fn += id2dam[id_][d]['neg'] if 'neg' in id2dam[id_][d].keys() else 0
            id2stats[id_] ={'fn':fn, 'fp':fp, 'tp':tp,
                         'precision':precision(tp, fp),
                         'recall':recall(tp, fn),
                         'f1': f_beta(tp, fp, fn,1)
                        }
    id2stats = {k: v for k, v in sorted(id2stats.items(), key=lambda item: item[1]['f1'])}
    return id2stats
```

```python
#get_d2stats
sp2dam = {}
for id_ in id2dam:
    sp, n = id_.split('/')
    if sp not in sp2dam.keys():
        sp2dam[sp]={}
        
    fp = tp = fn = 0
    for d in id2dam[id_].keys():
        if d not in sp2dam[sp].keys():
            sp2dam[sp][d] = {'pos':0, 'neg':0}
        sp2dam[sp][d]['pos'] += id2dam[id_][d]['pos']
        sp2dam[sp][d]['neg'] += id2dam[id_][d]['neg']
# get_id2stats(dam2stats)

dam2stats = {}
for sp in sp2dam.keys():
    dam2stats[sp]={}
    fp = tp = fn = 0
    for d in sp2dam[sp].keys():
        if d not in dam2stats[sp].keys():
            dam2stats[sp][d] = {}
        if d == '__background__':
            if 'fp' not in dam2stats[sp][d].keys():
                dam2stats[sp][d]['fp'] = 0
            dam2stats[sp][d]['fp'] += sp2dam[sp][d]['pos'] if 'pos' in sp2dam[sp][d].keys() else 0
        else:
            if 'tp' not in dam2stats[sp][d].keys():
                dam2stats[sp][d]['tp'] = 0
            if 'fn' not in dam2stats[sp][d].keys():
                dam2stats[sp][d]['fn'] = 0
            dam2stats[sp][d]['tp'] += sp2dam[sp][d]['pos'] if 'pos' in sp2dam[sp][d].keys() else 0
            dam2stats[sp][d]['fn'] += sp2dam[sp][d]['neg'] if 'neg' in sp2dam[sp][d].keys() else 0

tmp = dam2stats.copy() 
for sp in dam2stats.keys():
    s=dam2stats[sp]
    for d in s.keys():
        if d != '__background__':
            s[d]['recall']=recall(s[d]['tp'], s[d]['fn'])

tmp['tot']={}            
for sp in dam2stats.keys():
    s=dam2stats[sp]
    for d in s.keys():
        if d not in tmp['tot'].keys():
            tmp['tot'][d]={}
        for k in s[d].keys():
            if k not in tmp['tot'][d].keys():
                tmp['tot'][d][k] = 0
            tmp['tot'][d][k] += s[d][k]

for d in tmp['tot'].keys():
    if d == '__background__':
        continue
    tmp['tot'][d]['recall'] = recall(tmp['tot'][d]['tp'], tmp['tot'][d]['fn'])
    
df = {}
for sp in dam2stats.keys():
    s = dam2stats[sp]
    for d in s.keys():
        if d == '__background__':
            continue
        list_d=[round(v, 3) for k,v in s[d].items()]
        df[sp+'/'+d] = list_d
    
plt.figure(figsize=(10,5))
df = pd.DataFrame.from_dict(df)
colors = ['b','maroon','g']
# creating the bar plot
ori_cols = df.columns

df.columns = ['qb/ho', 'qb/mar', 'qb/sk', 'os/ho','os/mar','os/sk','ql/ho','ql/mar','ql/sk']
for i in range(3):
    plt.bar(df.columns[i*3:i*3+3], df.iloc[2,i*3:i*3+3],color=colors[i], width = 0.4)
    plt.ylabel('recall')
    plt.xlabel('species/damage')
df.index=['tp','fn','recall']
df
```

```python
def plt_(list_s, thre, id2paths,root_leaves,thres_num=10):
    def add_grid(img, t_size):
        m,n = img.shape[0], img.shape[1]
        for r in np.arange(0,m,t_size):
            img[r:r+2, :] = [0,0,255]
        for c in np.arange(0,n,t_size):
            img[:, c:c+2] = [0,0,255]
        return img
    for idx, id_ in enumerate(list_s.keys()):
        if idx >= thres_num:
            break
        if id_!='quercus-bicolor-herbivory/00000415_4287841': # id_ != 'quercus-bicolor-herbivory/00000719_2023078' and id_ != 'quercus-bicolor-herbivory/00000652_1305366':
            continue
        sp,n = id_.split('/')
        act = np.load(id2paths[id_]['act'])
        prods = sigmoid(act)
        preds = prods > 0.5
        img,_ = read_image(root_leaves/id2paths[id_]['image'])
        label,_=read_image(root_leaves/id2paths[id_]['label'])
        diff = np.absolute(preds - label)
        img_gt = img.copy()
        img_gt[label!=0] = [0, 255, 0]
        img_pred = img.copy()
        img_pred[preds!=0] = [255,0,0]
        tp = np.logical_and(label!=0, preds!=0)
        fp = np.logical_and(label==0, preds!=0)
        fn = np.logical_and(label!=0, preds==0)
        img_over = img.copy()
        
        img_over[fp]= [255,0,0]
        img_over[fn]= [0,0,255]
        img_over[tp]= [0,255,0]
        
        image_datas = [img, img, img_over, img_pred, preds]
        try_create_dir('./imgs')
        save_image(f'./imgs/original{n}.jpeg', img)
        save_image(f'./imgs/pred{n}.jpeg', img_over)
        
        t_size=512
        plt.figure(figsize=(25,17))
#         plt.subplot(1,3,1)
#         plt.imshow(add_grid(image_datas[0], t_size))
#         plt.xlabel(id_)
        plt.subplot(1,2,1)
        plt.imshow(add_grid(image_datas[1], t_size)) 
        plt.xlabel('gt--'+id_)
        plt.subplot(1,2,2)
        plt.imshow(add_grid(image_datas[2], t_size))
        plt.xlabel('pred--'+ id_)
        plt.show()
thre = 0.2
list_s = {k:v[score] for k,v in id2stats.items() if v[score]<=thre}
plt_(list_s, thre,id2paths,root_leaves)
```
